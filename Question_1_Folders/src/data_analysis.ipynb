```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv(r"C:\Users\User\Downloads\FALL_SEM\PDS\Question_1_Folders\data_clean\cleaned_frailty_data.csv")

df = pd.DataFrame(data)

# Calculate the correlation matrix with numeric_only=True
corr_matrix = df.corr(numeric_only=True)


# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title("Correlation Heatmap")


plt.savefig('C:/Users/User/Downloads/FALL_SEM/PDS/Question_1_Folders/results/heatmap.png')
# Show the plot
plt.show()


```
