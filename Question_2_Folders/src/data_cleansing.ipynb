```python
import pandas as pd

# Create a DataFrame with the provided data
data = "C:/Users/User/Downloads/FALL_SEM/PDS/Question_2_Folders/data_raw/StudentsPerformance_raw.csv"
df = pd.read_csv(data)

# Now, let's clean the data
# Convert 'gender' column to lowercase
df['gender'] = df['gender'].str.lower()

# Convert 'race/ethnicity' column to lowercase
df['race/ethnicity'] = df['race/ethnicity'].str.lower()

# Convert 'parental_level_of_education' column to lowercase
df['parental level of education'] = df['parental level of education'].str.lower()

# Convert 'lunch' column to lowercase
df['lunch'] = df['lunch'].str.lower()

# Convert 'test_preparation_course' column to lowercase
df['test preparation course'] = df['test preparation course'].str.lower()

# Check for any missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)

# Check for duplicate rows and remove them
df = df.drop_duplicates()

# Reset the index after removing duplicates
df = df.reset_index(drop=True)

# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df)

Cleaned_data = "C:/Users/User/Downloads/FALL_SEM/PDS/Question_2_Folders/data_clean/StudentsPerformance_cleaned.csv"

# Export the cleaned DataFrame to a CSV file
df.to_csv(Cleaned_data, index=False)

print("Cleaned data has been exported to:", Cleaned_data)
```
